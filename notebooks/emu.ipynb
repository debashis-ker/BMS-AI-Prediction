```python
# Process each equipment ID separately
equipment_ids = emu_df['equipment_id'].unique()
eg_stats = []

for eq_id in equipment_ids:
    # Filter data for this equipment
    eq_df = emu_df[emu_df['equipment_id'] == eq_id].copy()
    
    # Pivot the data
    pivoted = eq_df.pivot_table(
        index='timestamp',
        columns='datapoint',
        values='monitoring_data',
        aggfunc='first'
    )
    
    # Convert all columns to numeric
    for col in pivoted.columns:
        pivoted[col] = pd.to_numeric(pivoted[col], errors='coerce')
    
    # Round to 2 decimal places
    pivoted = pivoted.round(2)
    
    # Drop rows with any NA
    pivoted_clean = pivoted.dropna()
    
    # Get min and max for Eg if it exists
    if 'Eg' in pivoted_clean.columns:
        eg_stats.append({
            'Equipment_ID': eq_id,
            'min_Eg': pivoted_clean['Eg'].min(),
            'max_Eg': pivoted_clean['Eg'].max(),
            'mean_Eg': pivoted_clean['Eg'].mean(),
            'std_Eg': pivoted_clean['Eg'].std(),
            'count': pivoted_clean['Eg'].count(),
            'rows_before_dropna': len(pivoted),
            'rows_after_dropna': len(pivoted_clean)
        })

# Create summary DataFrame
eg_summary = pd.DataFrame(eg_stats).set_index('Equipment_ID')
print("Eg Statistics by Equipment ID:")
print(eg_summary)
```

```python
# Drop rows with any NA values
pivoted_table_clean = pivoted_table.dropna()

print(f"Shape after dropping NAs: {pivoted_table_clean.shape}")
print(f"Rows dropped: {pivoted_table.shape[0] - pivoted_table_clean.shape[0]}")
```

```python
# Get min and max "Eg" for each equipment_id
# Method 1: Using column slicing with xs (cross-section)
eg_stats = pd.DataFrame()

for equipment_id in pivoted_table_clean.columns.get_level_values(0).unique():
    if (equipment_id, 'Eg') in pivoted_table_clean.columns:
        eg_data = pivoted_table_clean[(equipment_id, 'Eg')]
        eg_stats.loc[equipment_id, 'min_Eg'] = eg_data.min()
        eg_stats.loc[equipment_id, 'max_Eg'] = eg_data.max()
        eg_stats.loc[equipment_id, 'mean_Eg'] = eg_data.mean()
        eg_stats.loc[equipment_id, 'std_Eg'] = eg_data.std()

print("Eg Statistics by Equipment ID:")
print(eg_stats)
```

```python
# Alternative Method 2: Using groupby on column levels
# This is more elegant for multi-index columns
eg_columns = [col for col in pivoted_table_clean.columns if col[1] == 'Eg']

eg_stats_alt = pd.DataFrame({
    'Equipment_ID': [col[0] for col in eg_columns],
    'min_Eg': [pivoted_table_clean[col].min() for col in eg_columns],
    'max_Eg': [pivoted_table_clean[col].max() for col in eg_columns],
    'mean_Eg': [pivoted_table_clean[col].mean() for col in eg_columns],
    'count': [pivoted_table_clean[col].count() for col in eg_columns]
}).set_index('Equipment_ID')

print("\nAlternative approach - Eg Statistics:")
print(eg_stats_alt)
```

```python
import matplotlib.pyplot as plt
df['PowSum'] = df['PowA'] + df['PowB'] + df['PowC']
df['PowDiff'] = df['PowT'] - df['PowSum']

print("Power Sum Check (Head):")
print(df[['PowA', 'PowB', 'PowC', 'PowSum', 'PowT', 'PowDiff']].head())
print(f"Mean Absolute Error in Power Sum: {df['PowDiff'].abs().mean()}")


df['VoltLN_Avg'] = (df['VoltAN'] + df['VoltBN'] + df['VoltCN']) / 3
df['VoltLL_Avg'] = (df['VoltBC'] + df['VoltCA']) / 2
df['VoltRatio'] = df['VoltLL_Avg'] / df['VoltLN_Avg']

print("\nVoltage Ratio Check (Head):")
print(df[['VoltLN_Avg', 'VoltLL_Avg', 'VoltRatio']].head())
print(f"Mean Voltage Ratio: {df['VoltRatio'].mean()}")
print(f"Expected Ratio: {np.sqrt(3)}")


df['PowA_calc'] = (df['VoltAN'] * df['CurrA'] * df['Pf']) / 1000
df['PowA_Error'] = df['PowA'] - df['PowA_calc']

print("\nPower A Calculation Check (Head):")
print(df[['VoltAN', 'CurrA', 'Pf', 'PowA', 'PowA_calc', 'PowA_Error']].head())


df = df.sort_values('timestamp')

df['TimeDiff_h'] = df['timestamp'].diff().dt.total_seconds() / 3600

df['PowT_Prev'] = df['PowT'].shift(1)
df['Energy_Calc_Delta'] = ((df['PowT'] + df['PowT_Prev']) / 2) * df['TimeDiff_h']

initial_Eg = df['Eg'].iloc[0]
df['Eg_Calc'] = df['Energy_Calc_Delta'].cumsum().fillna(0) + initial_Eg

print("\nEnergy Accumulation Check (Head):")
print(df[['timestamp', 'PowT', 'TimeDiff_h', 'Eg', 'Eg_Calc']].head())

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(df['PowT'], label='Measured PowT', color='blue', alpha=0.7)
plt.plot(df['PowSum'], label='Sum (A+B+C)', color='orange', linestyle='--', alpha=0.7)
plt.title('Proof 1: Total Power = Sum of Phases')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(df['VoltRatio'], label='Measured Ratio (LL/LN)', color='green')
plt.axhline(y=np.sqrt(3), color='red', linestyle='--', label='Theoretical sqrt(3)')
plt.title('Proof 2: Voltage Relationship (Sqrt(3))')
plt.legend()
plt.ylim(1.7, 1.76)
plt.grid(True)

plt.subplot(2, 2, 3)
plt.scatter(df['PowA_calc'], df['PowA'], alpha=0.5)
plt.plot([0, df['PowA'].max()], [0, df['PowA'].max()], 'r--')
plt.xlabel('Calculated Power A (V*I*Pf)')
plt.ylabel('Measured Power A')
plt.title('Proof 3: Power Calculation Formula')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(df['timestamp'], df['Eg'], label='Meter Energy Reading', linewidth=2)
plt.plot(df['timestamp'], df['Eg_Calc'], label='Calculated from Power', linestyle='--', color='red')
plt.title('Proof 4: Energy is Integral of Power')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
```
